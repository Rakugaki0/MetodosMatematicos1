```python
def pp(f, g):
    return integrate(f*g, (x, -1, 1))

def pp2(f,g, alg = "maxima"):
    return integral(f*g*sqrt(1-x**2), (x, -1, 1), algorithm=alg)

def npp2(f,g):
    return numerical_integral(f*g*sqrt(1-x**2),-1, 1)[0]
```

```python
pp(x**2,x**4) #La base no es ortogonal
```

```python
def orto(basis, p = pp, normalized = True): 
    #Funcion que recibe una base y una definici√≥n de producto interno
    #devuelve base ortonormal
    orth_basis = []

    for q in basis:
        for prev_q in orth_basis:
            q = q - p(prev_q, q)/p(prev_q, prev_q)*prev_q
        if normalized:
            q = q / sqrt(p(q, q))
        orth_basis.append(q)
    
    return orth_basis
```

```python
legendre = vector(orto(mono, normalized = False))
cheb = vector(orto(mono, pp2, normalized= False))
```

### Aproximaci√≥n de la funcion 

$h(x) = {\left(-x^{2} + 1\right)} \sin\left(3 \, x\right)$ en las diferentes bases.

```python
h(x) = sin(3*x)*(1-x**2)
```

```python
def expancion(pol, base, p = pp):
    result = []
    for e in base:
        result.append(p(e, h))
    return result
```

```python
h_m = vector(expancion(h(x), mono))
h_l = vector(expancion(h(x), legendre))
h_c = vector(expancion(h(x), cheb, p = pp))
```

### Error en funci√≥n del grado de la aproximaci√≥n.

```python
def err(f,g):
    return sqrt(numerical_integral(abs(f-g)**2,-1, 1)[0])
```

```python
err(h(x), h_l * legendre), err(h(x), h_m * mono), err(h(x), h_c * cheb)
```

```python
legen_grad = []
chev_grad = []
mon_grad = []
for i in range(1, 10 + 1):
    mon_grad.append(err(h(x), h_m[0:i] * cheb[0:i]))
    chev_grad.append(err(h(x), h_c[0:i] * cheb[0:i]))
    legen_grad.append(err(h(x), h_l[0:i] * legendre[0:i]))
```

```python
plt.title("Legendre")
td = np.arange(1, 11)
plt.scatter(td, np.array(legen_grad) * 100)
plt.plot(td, np.array(legen_grad) * 100)


plt.ylabel("Error [%]")
plt.xlabel("Grado")
```
